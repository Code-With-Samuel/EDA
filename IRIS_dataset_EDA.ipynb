```python
df.hist(figsize=(10, 6), bins=15, edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=15)
plt.show()
```

```python
df.hist(
    column=None,          # Specify a specific column or list of columns
    by=None,              # Group by a categorical column
    grid=True,            # Show grid (default is True)
    xlabelsize=None,      # X-axis label font size
    ylabelsize=None,      # Y-axis label font size
    bins=15,              # Number of bins (default is 10)
    figsize=(10, 6),      # Figure size (width, height)
    color="blue",         # Histogram color
    alpha=0.75,           # Transparency (0 = transparent, 1 = solid)
    edgecolor="black",    # Color of bin edges
    rwidth=0.9,           # Width of the bars relative to bin size
    log=False,            # Log scale on Y-axis
    layout=(2, 2),        # Arrange multiple histograms in a grid (rows, cols)
    sharex=False,          # Share X-axis across subplots
    sharey=False,          # Share Y-axis across subplots
    density=False,        # Show density instead of counts
    cumulative=False     # Show cumulative distribution
)
plt.suptitle("Feature Distributions", fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Boxplot of Features")
plt.show()
```

```python
df.boxplot(
    column=None,         # Specify columns to plot (default: all numeric columns)
    by=None,             # Group by a categorical column
    fontsize=None,       # Font size for axis labels
    grid=True,           # Show grid (default: True)
    rot=0,               # Rotate x-axis labels (0 = horizontal)
    figsize=(10, 6),     # Figure size (width, height)
    vert=True,           # Vertical (True) or Horizontal (False) boxplot
    showfliers=True,     # Show outliers (default: True)
    showmeans=False,     # Show mean in the boxplot
    patch_artist=True,   # Fill boxes with color
    notch=False,         # Use notched boxplot (shows confidence intervals)
    widths=0.7,         # Width of the boxes
    meanline=False,      # Show mean as a line instead of a dot
    showcaps=True,       # Show caps (top and bottom horizontal lines)
    whis=1.5            # Determines the range for whiskers
)
plt.suptitle('Boxplots', fontsize = 15)
plt.show()
```

```python
sns.pairplot(df, hue="species", markers=["o", "s", "D"])
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df_new=df.drop(columns=['species'])
sns.heatmap(df_new.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
